#IMPORTING

```python
import numpy as np
import pandas as pd
import itertools

from sklearn import datasets
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from mlxtend.plotting import plot_decision_regions
from mlxtend.preprocessing import shuffle_arrays_unison
import seaborn as sns

from tqdm.notebook import tqdm_notebook as tqdm

from sklearn.svm import SVC
```

#Aggregation

```python
colnames=["F1","F2","type"]
Aggregation_df = pd.read_csv("/content/drive/MyDrive/DATASET/ML DATA/Aggregation.txt", skiprows=7, sep='\t', names=colnames)
```

```python
Y=Aggregation_df["type"]
X=Aggregation_df.drop(columns=["type"])
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y ,test_size=0.2, random_state=42)
```

```python
svm = SVC(C=100000, kernel='rbf')
svm.fit(X_train, Y_train)
```

```python
# Plotting decision regions
plot_decision_regions(np.array(X_test), np.array(Y_test), clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('SVM on Aggregation')
plt.show()
accuracy = svm.score(X_test, Y_test)
print ("ACCURACY : ",accuracy*100," % ")
```

#IRIS

```python
#colnames=["F1","F2","type"]
iris_df = pd.read_csv("/content/drive/MyDrive/DATASET/ML DATA/Iris.csv")
```

```python
iris_df.columns
```

```python
for i in range(len(iris_df)):
  if iris_df["Species"][i]=="Iris-setosa":
    iris_df['Species'][i]=0

  elif iris_df["Species"][i]=="Iris-versicolor":
    iris_df['Species'][i]=1
  else :
    iris_df['Species'][i]=2
```

```python
iris_df['Species'] = iris_df['Species'].astype(int)
```

```python
iris_df.drop(columns="Id",inplace=True)
```

```python
Y=iris_df["Species"]
X=iris_df.drop(columns=["Species"])
```

```python
X=X.iloc[:, :2]
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y ,test_size=0.2, random_state=42)
```

```python
svm = SVC(C=900000, kernel='rbf')
svm.fit(X_train, Y_train)
```

```python
# Plotting decision regions
plot_decision_regions(np.array(X_test), np.array(Y_test), clf=svm, legend=2)

# Adding axes annotations

plt.title('SVM on Aggregation')
plt.show()
accuracy = svm.score(X_test, Y_test)
print ("ACCURACY : ",accuracy*100," % ")
```
